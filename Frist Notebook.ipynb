### Replacing null values in categorial featreus with the mode


```python
# I want to built a func that takes onlu categorical features and replace null values with the mode
```

```python
def fill_mode_in_cf(df):
    "it takes categorial features and replace null values with the mode"
    df_copy = df.copy()
    cat_var = [key for key in dict(df_copy.dtypes) if dict(df_copy.dtypes)[key] in ['object']]
    for col in cat_var:
        if col != 'Platform':
            df_copy[col].fillna(df_copy[col].mode()[0], inplace=True)
    return df_copy.isnull().sum()
```
